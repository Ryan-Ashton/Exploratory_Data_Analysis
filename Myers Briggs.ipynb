```python
#Load libraries
import pandas as pd
import numpy as np
import seaborn as sns
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from scipy.misc import imread
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Get the data
df = pd.read_csv('mbti_1.csv', encoding='utf-8')
df.head()
```

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.countplot(df['type'])
```

```python
#Credit to DLao from Kaggle for this for loop

fig, ax = plt.subplots(len(df['type'].unique()), sharex=True, figsize=(15,10*len(df['type'].unique())))

k = 0
for i in df['type'].unique():
    df_2 = df[df['type'] == i]
    wordcloud = WordCloud().generate(df_2['posts'].to_string())
    ax[k].imshow(wordcloud)
    ax[k].set_title(i)
    ax[k].axis("off")
    k+=1
```
